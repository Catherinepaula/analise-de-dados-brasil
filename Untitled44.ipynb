```python
from bs4 import BeautifulSoup
import pandas as pd

# Novo caminho para o arquivo XML carregado no Colab
caminho_arquivo = 'Profissao Analista de dados M13 Exercicio original.xml'

# Abrir e ler o arquivo
with open(caminho_arquivo, 'r', encoding='utf-8') as file:
    soup = BeautifulSoup(file, 'lxml')

# Extrair os dados
estados = soup.find_all('estado')
dados_estados = []

for estado in estados:
    nome = estado.find('nome').text.strip()
    sigla = estado.find('sigla').text.strip()
    regiao = estado.find('regiao').text.strip()
    dados_estados.append([nome, sigla, regiao])

# Criar DataFrame e salvar como CSV separado por ';'
estados_df = pd.DataFrame(dados_estados, columns=['estado', 'sigla', 'região'])
estados_df.to_csv('estados-limpo.csv', sep=';', index=False)
```

```python
import pandas as pd

# Carregar o CSV original das cidades
cidades_df = pd.read_csv('Profissao Analista de dados M13 Exercicio cidades-bruto.csv')

# Filtrar apenas os dados do Censo 2010
cidades_2010_df = cidades_df[cidades_df['Censo'] == 2010]

# Selecionar e renomear as colunas
cidades_2010_df = cidades_2010_df[['UF', 'nome', 'Pop_est_2009', 'PIB', 'PIB_percapita']]
cidades_2010_df.columns = ['estado', 'cidade', 'populacao', 'pib', 'pib_percapita']

# Salvar em CSV separado por ';'
cidades_2010_df.to_csv('cidades-limpo.csv', sep=';', index=False)
```

```python
import pandas as pd

# Carregar os arquivos limpos
estados_df = pd.read_csv('estados-limpo.csv', sep=';')
cidades_df = pd.read_csv('cidades-limpo.csv', sep=';')

# Padronizar os nomes dos estados (opcional, caso tenha divergência)
estados_df['estado'] = estados_df['estado'].str.upper().str.strip()
cidades_df['estado'] = cidades_df['estado'].str.upper().str.strip()

# Realizar o merge usando a coluna 'estado'
brasil_df = pd.merge(cidades_df, estados_df, on='estado', how='left')

# Verificar se há valores nulos
print(brasil_df.isnull().sum())

# Salvar o resultado
brasil_df.to_csv('brasil.csv', sep=';', index=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Gráfico de barras: 10 cidades menos populosas
menos_pop = brasil_df.nsmallest(10, 'populacao')
plt.figure(figsize=(10,6))
sns.barplot(x='populacao', y='cidade', data=menos_pop, palette='viridis')
plt.title('10 cidades menos populosas do Brasil')
plt.xlabel('População')
plt.ylabel('Cidade')
plt.tight_layout()
plt.show()

# Gráfico de pizza: proporção da população por região
pop_por_regiao = brasil_df.groupby('região')['populacao'].sum()
plt.figure(figsize=(8,8))
plt.pie(pop_por_regiao, labels=pop_por_regiao.index, autopct='%1.1f%%', startangle=140)
plt.title('Proporção da População por Região')
plt.axis('equal')
plt.show()
```
